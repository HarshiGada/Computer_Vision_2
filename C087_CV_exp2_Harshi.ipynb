Aim: to identify edges using Difference of Guassian Filter and change parameters of Gaussian filter and observe the effect. Apply same technique on other image. Change threshold value to identify changes.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage import data
from skimage.color import rgb2gray
```

```python
#img = data.coffee()
img = cv2.imread('/content/1.jpg')
```

```python
# Convert Color image to gray image
img_g = rgb2gray(img)
```

```python
plt.imshow(img)
```

```python
plt.imshow(img_g, cmap = 'gray')
```

task 2: Blur the image using Gaussian filter of different values of standard deviations

```python
img_L_1 = cv2.GaussianBlur(img_g, (19,19), 4)
img_L_2 = cv2.GaussianBlur(img_g, (19,19), 5)
img_L_3 = cv2.GaussianBlur(img_g, (19,19), 6)
img_L_4 = cv2.GaussianBlur(img_g, (19,19), 7)
img_H_5 = cv2.GaussianBlur(img_g, (19,19), 10)
```

```python
plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
plt.imshow(img_L_2, cmap = 'gray')
plt.title('Gaussian Filtered Image,Sigma=2')
plt.subplot(2,2,2)
plt.imshow(img_L_3, cmap = 'gray')
plt.title('Gaussian Filtered Image,sigma=3')
plt.subplot(2,2,3)
plt.imshow(img_L_4, cmap = 'gray')
plt.title('Gaussian Filtered Image,sigma=4')
plt.subplot(2,2,4)
plt.imshow(img_H_5, cmap = 'gray')
plt.title('Gaussian Filtered Image,sigma=10')
```

```python
# determine the difference of gaussian filter
dog1_10=img_L_1-img_H_5
dog2_10=img_L_2-img_H_5
dog3_10=img_L_3-img_H_5
dog4_10=img_L_4-img_H_5
```

```python
plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
plt.imshow(dog1_10, cmap = 'gray')
plt.title('Difference of Guassian Filtered Image,Sigma=1')
plt.subplot(2,2,2)
plt.imshow(dog2_10, cmap = 'gray')
plt.title('Difference of Guassian Filtered Image,sigma=2')
plt.subplot(2,2,3)
plt.imshow(dog3_10, cmap = 'gray')
plt.title('Difference of Guassian Filtered Image,sigma=3')
plt.subplot(2,2,4)
plt.imshow(dog4_10, cmap = 'gray')
plt.title('Difference of Guassian Filtered Image,sigma=4')
```

```python
th = np.max(dog1_10) * 0.05
dog1_10[np.abs(dog1_10) > th] = 255
dog2_10[np.abs(dog2_10) > th] = 255
dog3_10[np.abs(dog3_10) > th] = 255
dog4_10[np.abs(dog4_10) > th] = 255

plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
plt.imshow(dog1_10, cmap = 'gray')
plt.title('Difference of Guassian Filtered Image,Sigma=1')
plt.subplot(2,2,2)
plt.imshow(dog2_10, cmap = 'gray')
plt.title('Difference of Guassian Filtered Image,sigma=2')
plt.subplot(2,2,3)
plt.imshow(dog3_10, cmap = 'gray')
plt.title('Difference of Guassian Filtered Image,sigma=3')
plt.subplot(2,2,4)
plt.imshow(dog4_10, cmap = 'gray')
plt.title('Difference of Guassian Filtered Image,sigma=4')
```

Conclusion: The given image is blurred by gaussian filter with standard deviation of 1,2,3,4,10. The Difference of Gaussian is computed for following pair of standard deviations-

(1,10),

(2,10),

(3,10),

(4,10).

It is observed that for threshold of 20% of the maximum value of DoG(sigma=1,10), the number of edge pixels keep reducing with the increase in standard deviation from 1 to 4.If no sigma value is changed from 4 to 8, the value of DoG at most of the pixels go below threshold.Therefore, the threshold is reduced from 20% to 5% of maximum value.For other images noisy pixels like small edges can be removed by using higher value of low sigma value
If edges are thick size of filter can be lowered from 31,31 to lower filter size
